# Collaborative Filtering Based Recommender System

```python
Ratings_Name
```

```python
x = Ratings_Name.groupby('User-ID').count()['Book-Rating']>200
Rating_User = x[x].index
```

```python
Filter_Rating  = Ratings_Name[Ratings_Name['User-ID'].isin(Rating_User)]
```

```python
y = Filter_Rating.groupby('Book-Title').count()['Book-Rating']>=50
Books_Rating_50 = y[y].index
```

```python
Final_R = Filter_Rating[Filter_Rating["Book-Title"].isin(Books_Rating_50)]
Final_R
```

```python
Final_R.drop_duplicates()
```

```python
Final_R.duplicated().sum()
```

```python
pt = Final_R.pivot_table(index = 'Book-Title',columns = "User-ID",values = "Book-Rating")
```

```python
pt
```

```python
pt.shape
```

```python
pt.fillna(0,inplace = True)
```

```python
pt
```

# use Euclidean Series

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_Score = cosine_similarity(pt)
```

```python
Similarity_Score.shape
```

```python
#np.where(pt.index == "1984")[0][0]
```

```python

def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    Similar_item = sorted(list(enumerate(Similarity_Score[index])),key = lambda x: x[1],reverse =True)[1:6]
    
    data = []
    for i in Similar_item:
        item = []
        temp_df = Books[Books['Book-Title']== pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)
        
    return data       
```

```python
Movie = input("Enter the Movie Name--  ")
print("Recommended Movie For You ->\n")
print(recommend(Movie))
```
